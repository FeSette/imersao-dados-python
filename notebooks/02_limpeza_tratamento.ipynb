```python
import numpy as np

# Criação de um DataFrame de exemplo
df_salarios = pd.DataFrame({
    "nome": ["Any", "Felipe", "Sergio", "Gabriela", "Jussara"],
    "salario": [4000, np.nan, 5000, np.nan, 100000]
})

# Substitue os NULOS pelas médias com arredondamento
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))

# Substitue os NULOS pela mediana dos salários
df_salarios["salario_mediana"] = df_salarios["salario"].fillna(df_salarios["salario"].median())

df_salarios
```

```python
df_temperatura = pd.DataFrame({
    "dia": ["segunda", "terca", "quarta", "quinta", "sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperatura["preenchido_ffill"] = df_temperatura["temperatura"].ffill()

df_temperatura["preenchido_bfill"] = df_temperatura["temperatura"].bfill()


df_temperatura
```

```python
df_cidades = pd.DataFrame ({
    "nome": ["Any", "Felipe", "Sergio", "Gabriela", "Jussara"],
    "cidade": ["São Paulo", np.nan, "Sorocaba", np.nan, "Campinas"]
})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Não informado")

df_cidades
```
